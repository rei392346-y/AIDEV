## パーセプトロン

### 二値分類の準備

````python
# 「setosaかそれ以外か」の列
""" datasetに花の種類がSetosaかそれ以外かの列を追加 """
is_setosa = (dataset['species'] == 0).astype(np.int64)

# datasetに追加
dataset['is_setosa'] = is_setosa
dataset.head()
````

````python
# setosaとそれ以外で分類した散布図
"""
次に、petal_length, petal_width でsetosaを分類できるか、グラフにして確認してみましょう。
seabornの lmplot を使うと、色分けした散布図を作成できます。
"""
sns.lmplot(x='petal_width', y='petal_length', hue='is_setosa', data=dataset, fit_reg=False)
````

### パーセプトロンを使ってみよう

````python
# 必要なライブラリの追加読み込み
from sklearn.linear_model import Perceptron
from sklearn.model_selection import train_test_split
````

````python
"""今回は petal_width と petal_length を使って is_setosa を予測します。"""
# 目的変数（Y）：is_setosa、説明変数（X）：petal_widthとpetal_length
Y = np.array(dataset['is_setosa'])
X = np.array(dataset[['petal_width', 'petal_length']])

print("Y=",Y)
print("-" * 50)
print("X=",X)
````

````python
# 形状を確認
print("Y=", Y.shape, ", X=", X.shape)
````

````python
"""データの分割はホールドアウト法で行います。 比率は7:3"""
# データの分割
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=0)
X_train, X_valid, Y_train, Y_valid = train_test_split(X_train, Y_train, test_size=0.3, random_state=0)
````

````python
# パーセプトロンのモデルを作成し、学習して予測
# 目的変数（Y）：is_setosa（setosaかどうか？）、説明変数（X）：petal_widthとpetal_length
""" petal length (cm)：内花被片（内側の花びら）の長さ
    petal width (cm)：内花被片の幅 """
perceptron_model = Perceptron()
perceptron_model.fit(X_train, Y_train) # XとYの訓練データ
Y_pred = perceptron_model.predict(X_valid) # Xの検証データから、予測値を算出
````

````python
# 正解（Yの検証データの目的変数）とYの予測値（setosaかどうか？）との比較
print(Y_valid[:5])
print(Y_pred[:5])
````

### モデルの確認

````python
# 重みの確認
w0 = perceptron_model.intercept_[0] # 切片（バイアス）。もともとの「しきい値 θ」を重みに吸収したもの
w1 = perceptron_model.coef_[0][0] # 特徴 x1（例：petal_width） に掛かる重み。
w2 = perceptron_model.coef_[0][1] #特徴 x2（例：petal_length） に掛かる重み。

print("w0:", w0, ", w1:", w1, ", w2:", w2)
````

````python
# モデルの式
x1 = np.linspace(0, 3, 10)
print("x1.shape=",x1.shape)
print("x1=",x1)
x2 = (2 - 1.2 * x1) / 0.7
print("x2.shape=",x2.shape)
print("x2=",x2)

# 実際の値（setosa以外）　青の点
data = dataset.query("is_setosa == 0")
plt.scatter(data['petal_width'], data['petal_length'], color="blue", label="0")

# 実際の値（setosa)　オレンジの点
data = dataset.query("is_setosa == 1")
plt.scatter(data['petal_width'], data['petal_length'], color="orange", label="1")

# モデルの式（setosaの部分） 緑の線部分
plt.plot(x1, x2, color="green", linewidth=2)
plt.fill_between(x1, x2, color='green', linewidth=0, alpha=.05)

# グラフ全体の大きさを指定
plt.xlim(0, 2.8)
plt.ylim(0, 7.5)

# 凡例の表示
plt.legend(loc="upper left", title="is_setosa")
````

※前章で、重みは「w0: 2.0 , w1: -1.2 , w2: -0.7」と確認済み

この図を⾒ると、パーセプトロンで求めた重みが係数と切⽚になる直線は、setosaとそれ以外を分ける直線であることがわかります。

Sのx1，x2に、この直線より上に位置する点の座標の値を⼊れると、負になることが確かめられます。たとえば、 `(0.5, 3)` の点で試してみましょう。
```
s=2-1.2*0.5-0.7*3=-0.7
→ s=2-0.6-2.1=-0.7
s=w0+(w1*0.5)+(w2*3)=-0.7
```
この場合、パーセプトロンの出⼒は`0`となり、petal_width，petal_lengthが `(0.5, 3)` の値を取るアヤメはsetosaではないと判断されたことになります。

同様に、直線の下に位置する点の座標を⼊れると正になることが確かめられます。`(0.2, 2)` で試してみましょう。
```
s=2-1.2*0.2-0.7*2=0.36
→ s=2-0.24-1.4=0.36
s=w0+(w1*0.2)+(w2*2)=0.36
```
この場合はパーセプトロンの出⼒は`1`となり、petal_width，petal_lengthが `(0.2, 2)` の値を取るアヤメはsetosaと判断されたことになります．

### モデルの評価

````python
# train（学習データ）、valid（検証データ）、test（テストデータ）、pred（予測結果）
# 正解率
"""
正解率は「全体のなかで正解したものの割合」です。以降の説明では分母をオレンジ色の枠内、分子を緑の枠内で表します。
分母＝全体
分子＝正解したもの
"""
from sklearn.metrics import accuracy_score

accuracy_score(Y_valid, Y_pred) 
````

````python
# F1値
from sklearn.metrics import f1_score
"""
F1値（F値）は、適合率と再現率の調和平均（逆数の平均を求め、それを逆数にした値）です。
F1値は scikit-learn の f1_score で求められます。
"""
f1_score(Y_valid, Y_pred)
````

## ロジスティック回帰
パーセプトロンで「1か0」を分類できるようになりました。しかしこのままでは多クラス分類に対応できません。そこで機械学習のモデルの結果を「1か0」ではなく「setosaである確率」のように確率で出力する方法を考えてみましょう。

### ロジスティック回帰の目的関数
線形回帰モデルの目的関数は誤差の2乗の合計を最小化するものでした。
  （Σ（シグマ）の式）

ロジスティック回帰の出力は確率なので、目的関数も `「分類が正解となる確率」を最大化する` ものとします。すべての学習データの事象が独立していると考えると、目的関数の数式は以下のようになります。これを `尤度関数` と呼びます。尤度とは「もっともらしさ（`尤もらしさ）`、どのくらい理にかなっているか」という意味です。Lは尤度の英語表記「Likelihood」の頭文字です。
　（Π（パイ）の式）
ただし、 fθ(x) は分類ごとの条件付き確率となります。たとえば0か1の二値分類であれば、すべての学習データにおいて以下をかけ合わせたものとなります。
- 目的関数=1の場合  となる確率
- 目的関数=0の場合  となる確率
尤度関数は確率の掛け合わせなので、値がものすごく小さくなります。そこで実際に計算する際は対数を取った `対数尤度関数` が良く使われています。

### 対数尤度と交差エントロピー
対数尤度関数は「分類が正解となる確率」を最大化するものなので、最適値も最大値を求めることとなります。同様の指標で `交差エントロピー`（crossentropy、交差エントロピー誤差）というものもあります。交差エントロピーとは2つの確率分布の距離を表したものです。目的変数と予測値との距離のようなものになります。

もっと簡単に言うと、実は、交差エントロピーは対数尤度関数にマイナスを掛けたものになります。対数尤度を最大化することは、交差エントロピーを最小化することと同じです。交差エントロピーを用いるのは、最適化問題として最小化で考えたほうが解きやすい場合です。

具体的な計算式までは理解しなくても問題ありませんので `交差エントロピー` という用語を覚えておくと良いでしょう。

ロジスティック回帰では多クラス分類を行うことができます。モデルの出力は 各クラスに属する確率 となるように拡張されています。このように「各クラスに属する確率」を表現できるように拡張した関数を ソフトマックス関数（softmax）と呼びます。ソフトマックス関数は、ディープラーニングの分類モデルなど、機械学習で多く用いられている関数の1つです。

それでは多クラス分類を行ってみましょう。ここではすべての説明変数を使用して`アヤメの品種（species）`を予測してみます。

````python
# Y:目的変数、X:説明変数
Y = np.array(dataset['species']) # アヤメの品種
print("Y.shape=",Y.shape)
print("Y=",Y)
print("-"*50)
"""
sepal length (cm)：外花被片（外側の花びら）の長さ
sepal width (cm)：外花被片の幅
petal length (cm)：内花被片（内側の花びら）の長さ
petal width (cm)：内花被片の幅
"""
X = np.array(dataset[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']])
print("X.shape=",X.shape)
print("X=",X)
````

## 決定木
決定木 (decision tree)とはデータを分割して分類を判定するモデルです

### 決定木モデルの説明 6
データ（散布図）の領域の分割を繰り返していくと以下のようになります  

  
この番号を使って、分割の過程をグラフにすると、以下のようになります。この形が木のように見えることから決定木と呼ばれています。

### 決定木モデルの説明 7
  

決定木の途中の丸を ノード と呼びます。末端のノードは リーフ と呼ばれます。ノードとノードを繋ぐ線のことを リンク や エッジ と呼びます。決定木は判断の課程がわかりやすいため、機械学習だけでなく意思決定やデータの可視化の手法として広く用いられています。

決定木は、データのまとまりがなるべく単一のカテゴリーになるように、データを分割します。データに他のカテゴリーの値が入っている割合を 不純度 と呼びます。この不純度がなるべく小さくなるようにデータを分割します。グラフにすると以下のようなイメージです。なお、覚える必要はありませんが、不純度の計算にはジニ係数（ジニ不純度）や交差エントロピーといった計算式が使われます。

## 決定木を使ってみよう

````python
# 必要なライブラリの追加読み込み
"""
それでは決定木による分類モデルを作成してみましょう。
決定木は scikit-learn の DecisionTreeClassifier クラスを使います。
"""
from sklearn.tree import DecisionTreeClassifier
````

````python
"""
データは、ロジスティック回帰のところで作成したものをそのまま使います。
決定木のモデルを作成し、学習と予測、評価を行います。
下記のコードでは、データを分割する基準（不純度）の計算式に交差エントロピーを指定しています（criterion='entropy'）。
"""
# 決定木のモデルを作成し、学習して予測
tree_model = DecisionTreeClassifier(criterion='entropy')
tree_model.fit(X_train, Y_train)
Y_pred = tree_model.predict(X_valid)

# モデルの評価
print(classification_report(Y_valid, Y_pred))
````

````python
"""
これまでのモデルより少し性能が悪いようです。
前のレッスンで学んだグリッドサーチを使ってパラメータをチューニングしてみましょう。
"""
# グリッドサーチを使う
from sklearn.model_selection import GridSearchCV

# モデル:決定木
model = DecisionTreeClassifier()

"""
'gini' と 'entropy' は、決定木が分割の良し悪しを測る“不純度(impurity)”の指標です。
'gini'はジニ不純度、'entropy'は（シャノン）エントロピー＝情報利得を使います。
max_depth は木の深さの上限、min_samples_split はノードをさらに分割するのに必要な最小サンプル数です
"""
# max_depth:木の深さ, min_samples_split:ノードを分割するために必要なデータの数
params = {
    # criterion=['gini','entropy'] はグリッドサーチで“どちらか”を試す候補を並べているだけで、同時使用ではありません。もちろん片方だけ指定も可能
    # 「探索の候補として['gini', 'entropy']の両方を試す」から。1本の木では常にどちらか一方しか使わない
    'criterion': ['gini', 'entropy'],
    'max_depth': [5, 10, 50],
    'min_samples_split': [2, 4, 10]
}

# グリッドサーチ
gscv = GridSearchCV(model, param_grid=params, cv=3)

# データの分割:機械学習用データを学習と検証に分けるのはクロスバリデーションで行ってくれる
Xg_train, Xg_test, Yg_train, Yg_test = train_test_split(X, Y, test_size=0.3, random_state=0)

gscv.fit(Xg_train, Yg_train)
````

````python
# 最適なパラメータ
"""
best_params(最適なパラメータのcriterion（分割基準）は 'gini'が選択された)
"""
gscv.best_params_
````

````python
# 最適なパラメータによる決定木のモデル作成と学習、予測
tree_model = DecisionTreeClassifier(criterion='gini', max_depth=10, min_samples_split=10)
tree_model.fit(X_train, Y_train)
Y_pred = tree_model.predict(X_valid)

# モデルの評価
print(classification_report(Y_valid, Y_pred))
````

## ランダムフォレスト
ランダムフォレストとは、複数の決定木を組み合わせて予測性能を高くしたモデルです。
- 複数の決定木モデルを用意する
- それぞれの決定木の学習データは、もとの学習データから重複を許して同じ数をランダムに抽出したものを使用する（決定木ごとに微妙に学習データを変えることで学習のバリエーションを増やす）
- 予測には、各決定木の予測結果の「多数決」を用いる
### アンサンブル学習
ランダムフォレストの考え方は「予測性能の低いモデル（決定木）を組み合わせて予測性能の高いモデルをつくる」というものです。このような方法を一般に、アンサンブル学習（ensemble learning）と呼びます。アンサンブルはフランス語で「（組み合わせた）全体」という意味です。合奏や合唱でも使われる言葉です。

代表的なアンサンブル学習の方法には、以下の3つがあります。
- バギング（bagging）：学習データ全体から重複を許してサンプリングを行い、多数の等価なモデルを 並列 に学習させて多数決を取る方法。過学習を防ぎつつ、正確性を上げるのに有効。 ランダムフォレスト はバギングに含まれる
- ブースティング（boosting）：多数の等価な単位モデルを 直列 につなげる方法。前のモデルの予測誤りの情報を後のモデルが受け取り、より精度の高い予測を行うようにする。ランダムフォレストに 勾配ブースティング (Gradient Boosting) という手法を組み合わせた XGBoost や LightGBM というモデルが知られている
- スタッキング（stacking）：前のモデルの予測結果を後のモデルの入力として使うのはブースティングと同様だが、多数の必ずしも等価でないモデルを組み合わせた手法。ブースティングのモデルをさらに組み合わせるなど、より複雑なモデルを構築できる

````python
# 必要なライブラリの追加読み込み
"""ランダムフォレストは scikit-learn の RandomForestClassifier クラスを使います。"""
from sklearn.ensemble import RandomForestClassifier
````

````python
"""データは、ロジスティック回帰のところで作成したものをそのまま使います。
また、「n_estimators」は木の数を指定するパラメータです。"""
# ランダムフォレストのモデルを作成し、学習して予測
randomforest_model = RandomForestClassifier(n_estimators=100)
randomforest_model.fit(X_train, Y_train)
Y_pred = randomforest_model.predict(X_valid)

# モデルの評価
print(classification_report(Y_valid, Y_pred))
````

## サポートベクターマシン（SVM）
まずは基本的な「線形で二値分類するモデル」を例にします。これは 線形SVM と呼ばれます。

SVMとは、マージンを最大化するようにデータを分割する境界線（この例では2次元のため直線、高次元では超平面と呼ばれます。図では赤い実線）を決定するモデルです。マージンとは図の「青い矢印」を指し、各データと分割する線（赤い実線）との距離を表します。ここで、分割線を決めるのに重要なのは 分割線から近いデータ です。この分割線から近いデータを サポートベクトル と呼びます。

サポートベクトルからのマージンが最大になるように分割線を決める、というのが基本的な考え方です。

**SVMのイメージ**  
  
SVMは分割境界を決定するためにサポートベクトルのみを考慮するため、特にデータが高次元であっても効率的に学習できる場合があり、計算量が比較的少ないという特徴につながることがあります。


### 非線形の分類への応用
線形SVMでは直線でデータを分割します。直線で分割できない 非線形の分類 へ対応するために、SVMではカーネル法（カーネルトリック）という手法で 学習データを変換してから モデルに学習させる、という方法を取ります。

### 多クラス分類への応用
複数のSVMを組み合わせることで多クラス分類モデルを実現できます。例として A, B, C の3つに分類したい場合は、以下のどちらかの方法となります。両者とも予測時にはそれらのモデルの多数決で分類を決めます。

- すべての組み合わせの3つのモデルを作成
  - 「A, B」,「A, C」,「B, C」
- 1つとそれ以外の組み合わせの3つのモデルを作成
  - 「A, BC」,「B, AC」,「C, AB」

## Lesson 8 Chapter 7.3 階層型クラスタリング
階層型クラスタリングは、ボトムアップでクラスタを作成するアルゴリズムです。
1. 各データ（ノード）間の距離を総当たりで調べる
2. もっとも距離が近いノードのペアをまとめて、1つのノード（クラスタ）と定義し直す。
3. 2でまとめられたノードを1つのノードと見て、その他のノードとの距離（リンケージと呼ばれる）を計算し直す。
4. すべてが1つのノードとしてまとまるまで、2と3を繰り返す。

階層型クラスタリングは、あらかじめクラスタ数を決める必要がないというメリットがあります。後述する樹形図（デンドログラム）を使い、クラスタリングの途中課程の任意の位置でクラスタ数を決めることができます。

````python
# 階層型クラスタリングを使ってみよう
"""まずは必要なライブラリをインポートします。"""
# 必要なライブラリの追加import
from scipy.spatial.distance import pdist, squareform
from scipy.cluster.hierarchy import dendrogram, linkage
````

````python
# 総当り距離の計算
"""
総当たり距離の計算は、 scipy の pdist() 関数を使用します。
距離の計算方法はデフォルトではユークリッド距離（euclidean）ですが、他のタイプの距離を用いることもできます。
"""
# 総当り距離の計算
dists = pdist(X)
dists.shape
````

````python
"""
distsは、「condensed form」と呼ばれる上三角部分（+α）を1次元配列の形で保存した行列表現になっています。
距離として確認するには、 scipy の squareform() メソッドを使用します。
"""
# 150 X 150 の行列の形に変換
dists2 = squareform(dists)
dists2.shape
````

````python
"""距離をヒートマップで可視化してみましょう。"""
# ヒートマップによる可視化
sns.heatmap(dists2)
````

色が濃いほど距離が近いことを表します。もともとのデータは品種ごとに並んでいたため、それぞれの品種ごとに距離が近い、すなわちクラスタとしてまとまっていることがわかります。

````python
# 階層型クラスタリングの実行
"""さて、距離が定義できたところで、クラスタリングを行います。階層型クラスタリングに対応する関数は、 scipy の linkage メソッドを使用します。"""
# 階層型クラスタリングの実行
"""
カリキュラムはdistsになっているが、ここはdists2じゃないとおかしい 
"""
Z = linkage(dists2, 'ward') 
````

第2引数（ここでは 'ward'）が、リンケージ（クラスタと別のノードとの距離）の計算アルゴリズムを表しています。他にも、第2引数で指定できるアルゴリズムがありますので、以下にまとめます。
- ward：  
Ward’s minimum variance method と呼ばれる方法です。クラスタ内の「重心」を求め、重心と他のノードとの距離を計算します。他の方法と比較して、外れ値による影響が少ないため、多く用いられています

- single：  
もっとも単純なアルゴリズムです。「クラスタ内のノードのうち、対象ノードにもっとも近いノードの距離」がクラスタと対象ノードとの距離として計算されます。ただしこの場合クラスタが大きくなればなるほど、「本来の」距離に関係なく他のノードがそのクラスタに追加されやすくなってしまうという問題点があり、あまり利用されません。

- average：  
新しいクラスタの「中心」を構成ノードの内分点によって定義するものです。ただしこの方法は、クラスタ内の外れ値の影響で他のノードとの距離が、大きく算出される傾向があります。

````python
# Zの中身を全件確認
print("Z=",Z)
````

````python
# デンドログラムの作成
"""
最後にデンドログラムというグラフでクラスタリングの結果を確認しましょう。
デンドログラムは樹形図とも呼ばれます。作成は scipy の dendrogram メソッドを使用します。
"""
# グラフの大きさを指定
fig = plt.figure(figsize=(12, 12))

# X軸とY軸のラベルを指定
plt.xlabel('sample index')
plt.ylabel('distance')

# デンドログラムの作成
dendrogram(
    Z,
    leaf_rotation=90.,  # X軸のラベルを90度回転させる
    leaf_font_size=6.,  # X軸のラベルのフォントサイズを指定
)

# この行を入れないとdendrogramの中身が全て表示されてしまう。
plt.show()
````
